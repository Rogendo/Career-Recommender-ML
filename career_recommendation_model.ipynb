```python
# decision tree model

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X = pd.DataFrame(data.iloc[:,2:8])
```

```python
X
```

```python
y = data['Career Recommendation']
```

```python
y.isnull()
```

```python
y.shape
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
import pickle
from sklearn.preprocessing import MinMaxScaler


# create and fit the scaler
scaler = MinMaxScaler()
scaler.fit(X_train)

# transform the data using the scaler
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=20,criterion='gini', random_state=1, max_depth=3)
classifier.fit(X_train,y_train)
# classifier.fit(X_train_scaled,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# saving the model as a pkl file
import pickle
with open('trained_model.pkl', 'wb') as f:
    pickle.dump(classifier, f)
```

```python
# save the scaler as a file
with open('min_max_scaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
classifier.score(X_test, y_test)
```
